Question 2(a)

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义原函数 f(t) = |t|
def f(t):
    return np.abs(t)

# (a) 设置 m=5，生成5个等距节点
m = 5
x_nodes = np.linspace(-1, 1, m)
y_nodes = f(x_nodes)

# 拟合一个 m-1 = 4 次多项式插值函数
coeffs = np.polyfit(x_nodes, y_nodes, deg=m-1)

# 在区间 [-1, 1] 上生成1001个等距点用于画图
x_eval = np.linspace(-1, 1, 1001)
f_eval = f(x_eval)                         # 原函数在这些点的值
p_eval = np.polyval(coeffs, x_eval)        # 多项式插值函数在这些点的值
error = np.abs(f_eval - p_eval)            # 插值误差

# 绘制 f(t) 与 p(t) 的图像
plt.figure(figsize=(10, 5))
plt.plot(x_eval, f_eval, label='f(t) = |t|', linewidth=2)
plt.plot(x_eval, p_eval, label='Interpolating Polynomial (m=5)', linestyle='--')
plt.scatter(x_nodes, y_nodes, color='red', zorder=5, label='Interpolation Nodes')
plt.title('Function and Polynomial Interpolant (m=5)')
plt.xlabel('t')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# 绘制误差图像 |f(t) - p(t)|
plt.figure(figsize=(10, 5))
plt.plot(x_eval, error, color='purple')
plt.title('Interpolation Error |f(t) - p(t)| (m=5)')
plt.xlabel('t')
plt.ylabel('Error')
plt.grid(True)
plt.show()
```

Question 2(b)

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义原函数 f(t) = |t|
def f(t):
    return np.abs(t)

# 封装成函数以便多次调用
def plot_polyfit_interpolation(m):
    # 1. 生成 m 个等间距插值节点
    x_nodes = np.linspace(-1, 1, m)
    y_nodes = f(x_nodes)

    # 2. 用 polyfit 拟合 m-1 次多项式
    coeffs = np.polyfit(x_nodes, y_nodes, deg=m-1)

    # 3. 在1001个点上评估函数和插值多项式
    x_eval = np.linspace(-1, 1, 1001)
    f_eval = f(x_eval)
    p_eval = np.polyval(coeffs, x_eval)
    error = np.abs(f_eval - p_eval)

    # 4. 绘制 f(t) 与 p(t)
    plt.figure(figsize=(10, 5))
    plt.plot(x_eval, f_eval, label='f(t) = |t|', linewidth=2)
    plt.plot(x_eval, p_eval, label=f'Interpolating Polynomial (m={m})', linestyle='--')
    plt.scatter(x_nodes, y_nodes, color='red', zorder=5, label='Interpolation Nodes')
    plt.title(f'Function and Polynomial Interpolant (m={m})')
    plt.xlabel('t')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

    # 5. 绘制误差图像
    plt.figure(figsize=(10, 5))
    plt.plot(x_eval, error, color='purple')
    plt.title(f'Interpolation Error |f(t) - p(t)| (m={m})')
    plt.xlabel('t')
    plt.ylabel('Error')
    plt.grid(True)
    plt.show()

# 运行：m = 11
plot_polyfit_interpolation(11)

# 运行：m = 21
plot_polyfit_interpolation(21)
```

Question 2(c)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# 定义原函数 f(t) = |t|
def f(t):
    return np.abs(t)

# 封装函数 多次调用
def plot_cubic_spline_interpolation(m):
    # 1. 生成 m 个等间距节点
    x_nodes = np.linspace(-1, 1, m)
    y_nodes = f(x_nodes)

    # 2. 使用 CubicSpline 插值，默认 'not-a-knot' 条件
    spline = CubicSpline(x_nodes, y_nodes)

    # 3. 在 1001 个点上评估
    x_eval = np.linspace(-1, 1, 1001)
    f_eval = f(x_eval)
    s_eval = spline(x_eval)
    error = np.abs(f_eval - s_eval)

    # 4. 绘制 f(t) 与 s(t)
    plt.figure(figsize=(10, 5))
    plt.plot(x_eval, f_eval, label='f(t) = |t|', linewidth=2)
    plt.plot(x_eval, s_eval, label=f'Cubic Spline Interpolant (m={m})', linestyle='--')
    plt.scatter(x_nodes, y_nodes, color='red', zorder=5, label='Interpolation Nodes')
    plt.title(f'Function and Cubic Spline Interpolant (m={m})')
    plt.xlabel('t')
    plt.ylabel('Value')
    plt.legend()
    plt.grid(True)
    plt.show()

    # 5. 绘制误差图像
    plt.figure(figsize=(10, 5))
    plt.plot(x_eval, error, color='green')
    plt.title(f'Spline Interpolation Error |f(t) - s(t)| (m={m})')
    plt.xlabel('t')
    plt.ylabel('Error')
    plt.grid(True)
    plt.show()

# 运行：m = 5
plot_cubic_spline_interpolation(5)

# 运行：m = 11
plot_cubic_spline_interpolation(11)

# 运行：m = 21
plot_cubic_spline_interpolation(21)
```
